```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy
```

## Create random data from specific stats

## Calculate the cov from the stats above

```python
mean_wq = 1.3596107
std_wq = 0.224064
mean_flow = 2.969030
std_flow = 0.224120
```

For a covariance matrix [[6, -3], [-3, 3.5]]. The expected variances of the first and second components of the sample are 6 and 3.5, respectively, and the expected correlation coefficient is -3/sqrt(6*3.5) â‰ˆ -0.65465.

```python
corr = 0.6
x = 0.6 * np.sqrt(std_wq**2 * std_flow**2)
cov = [[std_wq**2, x], [x, std_flow**2]]
```

```python
cov
```

```python
pts = np.random.multivariate_normal([mean_wq, mean_flow], cov, size=1000)
pts = pd.DataFrame(pts, columns=["Quality", "Flow"])
```

## Visualise

```python
print(
    "Mean\n", pts.mean(axis=0),
    "\nStd\n", pts.std(axis=0), 
    "\nCovariance\n", np.cov(pts.T), 
    "\nCorrelation\n", np.corrcoef(pts.T)[0, 1]
)
```

```python
g = sns.jointplot(
    x="Quality",
    y="Flow",
    data=pts,
    kind="reg",
    truncate=False,
    height=5,
    scatter_kws={"alpha": 0.1},
)
```
